## Exercise 5



## Exercise 13
Find values of $x$ where $f(x)$ is continuous

$$
f(x) = 
\begin{cases}
6-2x & \hbox{ if $x<3$}\\
x^2-5x+5 & \hbox{ if $x \geq 3$}
\end{cases}.
$$

```{dropdown} Show answer
Answer: everywhere except $x = 3$
```



## Exercise 14
Find the value for $k$ that makes $f(x)$ continuous at $x=2$.

$$
f(x) =
\begin{cases}
kx^2+5 & \hbox{ if $x<2$}\\
21-kx -x^2 & \hbox{ if $x \geq 2$}
\end{cases}
$$

```{dropdown} Show answer
Answer: $k = 2$
```
